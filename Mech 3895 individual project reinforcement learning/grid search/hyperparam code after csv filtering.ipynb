```python
#What this code will do is read the filtered csv files that got scores of 500, and only
#use the hyperparameters that got to those scores






import gym
import random
import csv
import time
from itertools import product
from stable_baselines3 import PPO
from stable_baselines3.common.evaluation import evaluate_policy

hyperparams_list = []
with open('hyperparam_results_grid_search_filteredAAAAAAAARHGHGHGHGH at 6000 episodes.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        hyperparams_list.append(row)
best_hyperparams = None
best_score = -float('inf')

with open('Hyperparam results second test at 3000 episodes.csv', 'a', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['learning_rate', 'n_steps', 'batch_size', 'gamma', 'gae_lambda', 'clip_range', 'ent_coef', 'epsilon', 'mean_reward', 'training_time', 'standard deviation'])

    # this Loops over all combinations of hyperparameters in the csv file
    for hyperparams in hyperparams_list:
        learning_rate = float(hyperparams['learning_rate'])
        n_steps = int(hyperparams['n_steps'])
        batch_size = int(hyperparams['batch_size'])
        gamma = float(hyperparams['gamma'])
        gae_lambda = float(hyperparams['gae_lambda'])
        clip_range = float(hyperparams['clip_range'])
        ent_coef = float(hyperparams['ent_coef'])
        epsilon = float(hyperparams['epsilon'])

        # then the code does the same thing as before
        env = gym.make('CartPole-v1')

        model = PPO('MlpPolicy', env,
                    learning_rate=learning_rate,
                    n_steps=n_steps,
                    batch_size=batch_size,
                    gamma=gamma,
                    gae_lambda=gae_lambda,
                    clip_range=clip_range,
                    ent_coef=ent_coef,
                    clip_range_vf=epsilon,
                    verbose=1)

        print("Training with hyperparameters:", hyperparams)

        start_time = time.time()
        model.learn(total_timesteps=3000)
        end_time = time.time()

        training_time = end_time - start_time

        # this will evaulate the performance and see if we get a score of 500
        mean_reward, std_reward = evaluate_policy(model, env, n_eval_episodes=10)
        print("Mean reward: {:.2f} +/- {:.2f}".format(mean_reward, std_reward))

        if mean_reward > best_score:
            best_hyperparams = hyperparams
            best_score = mean_reward
            print("New best hyperparameters found:", best_hyperparams)
            print("New best score found:", best_score)


        with open('Hyperparam results second test at 3000 episodes.csv', 'a', newline='') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow((learning_rate, n_steps, batch_size, gamma, gae_lambda, clip_range, ent_coef, epsilon, mean_reward,training_time, std_reward))

    # Print the best hyperparameters and score (just for checking the code is running properly as it runs)
    print("Best hyperparameters:", best_hyperparams)
    print("Best score:", best_score)
```
