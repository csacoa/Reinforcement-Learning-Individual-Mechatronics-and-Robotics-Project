```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("CartPole-v1_PPO TESTING_AT_5_THOUSAND_LEARNING_RATE.csv")
df['Mean Score'] = pd.to_numeric(df['Mean Score'], errors='coerce')
mean_scores = df.groupby('Learning Rate')['Mean Score'].mean()
x = mean_scores.index
y = mean_scores.values
m, b = np.polyfit(x, y, 1)
plt.plot(x, y, 'o')
plt.plot(x, m*x + b)
plt.xlabel('Learning Rate')
plt.ylabel('Mean Score')
plt.title('Mean Score vs. Learning Rate')
#plt.savefig('mean_score_vs_learning_rate (5,000 episodes training).png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

df = pd.read_csv("CartPole-v1_PPO GAMMA TESTING AT 12500 EPISODES LEARNING RATE IS 0.001.csv")
df['Mean Score'] = pd.to_numeric(df['Mean Score'], errors='coerce')
mean_scores = df.groupby('Gamma')['Mean Score'].mean()
num_scores_equal_to_500 = len(df[df['Mean Score'] == 500])
num_total_scores = len(df['Mean Score'])
percent_scores_equal_to_500 = (num_scores_equal_to_500 / num_total_scores) * 100
plt.plot(mean_scores.index, mean_scores.values)
plt.xlabel('Gamma')
plt.ylabel('Mean Score')
plt.title('Mean Score vs. Gamma Trained at 12500 episodes Learning Rate = 0.001')

# Calculate the linear regression model and plot the line of best fit
slope, intercept, r_value, p_value, std_err = stats.linregress(mean_scores.index, mean_scores.values)
plt.plot(mean_scores.index, intercept + slope*mean_scores.index, 'r', label='Line of Best Fit')

plt.legend()

plt.show()

print(percent_scores_equal_to_500)
```
