```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("ALL DATA.csv")

acrobot_data = df[df["Environment"] == "Cart Pole"]

without_training_value = -500

algorithm_name = "A2C"  # Filter for PPO algorithm only

algorithm_data = acrobot_data[acrobot_data["Algorithm Name"] == algorithm_name]

colors = ['red', 'green', 'blue', 'purple', 'orange', 'black', 'pink', 'brown']  # Define colors for each point

key_labels = ['10,000 episodes', '20,000 episodes', '50,000 episodes', '100,000 episodes', '500,000 episodes', '1 million episodes', '10 million episodes']

for i in range(len(key_labels)):
    plt.scatter([], [], c=colors[i], alpha=1, s=50, label=key_labels[i], edgecolors='black')

plt.scatter(algorithm_data["Training time"], algorithm_data["Mean Score"], c=colors[:len(algorithm_data)], s=50, edgecolors='black', linewidth=1.5, zorder=2)
plt.plot(algorithm_data["Training time"], algorithm_data["Mean Score"], linewidth=2, color='gray', zorder=1)

plt.axhline(y=without_training_value, color='red', linestyle='--', label="Without Training")

plt.xlabel("Training time (s)")
plt.ylabel("Mean score")
plt.title("Cart Pole: Training time vs. Mean scores (PPO only)")
plt.legend(title="Number of episodes")


#plt.savefig("CartPole_training_time_vs_mean_score_ONLY PPO TO SEE DIMINISHING RETURNS.png")
plt.show()
```
