```python
import re

with open('hyperparam test 3', 'r') as f:
    lines = f.readlines()

# this extracts the numbers from each line using regular expressions
numbers = []
for line in lines:
    match = re.search(r'Score:\[(\d+\.\d+)\]', line)
    if match:
        numbers.append(match.group(1))

# this foormats the numbers as a comma-separated string
output = ','.join(numbers)

print(output)
```
