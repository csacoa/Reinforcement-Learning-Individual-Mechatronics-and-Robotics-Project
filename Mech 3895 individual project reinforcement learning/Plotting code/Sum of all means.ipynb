```python
import pandas as pd
import matplotlib.pyplot as plt

environments = ["Bipedal Walker","Pendulum", "Car Racing", "Mountain Car Continuous", "Lunar Lander", "Mountain Car" , "Cart Pole", "Acrobot", "Bipedal Hardcore"]

for env in environments:
    df = pd.read_csv("ALL DATA.csv")
    acrobot_data = df[df["Environment"] == env]
    ppo_sum = acrobot_data[acrobot_data["Algorithm Name"] == "PPO"]["Mean Score"].sum()
    td3_sum = acrobot_data[acrobot_data["Algorithm Name"] == "TD3"]["Mean Score"].sum()
    ddpg_sum = acrobot_data[acrobot_data["Algorithm Name"] == "DDPG"]["Mean Score"].sum()
    a2c_sum = acrobot_data[acrobot_data["Algorithm Name"] == "A2C"]["Mean Score"].sum()
    sac_sum = acrobot_data[acrobot_data["Algorithm Name"] == "SAC"]["Mean Score"].sum()
    dqn_sum = acrobot_data[acrobot_data["Algorithm Name"] == "DQN"]["Mean Score"].sum()
    labels = ["PPO", "TD3", "DDPG", "A2C", "SAC", "DQN"]
    scores = [ppo_sum, td3_sum, ddpg_sum, a2c_sum, sac_sum, dqn_sum]
    print("Sum of mean scores for PPO in", env, ":", ppo_sum)
    print("Sum of mean scores for TD3 in", env, ":", td3_sum)
    print("Sum of mean scores for DDPG in", env, ":", ddpg_sum)
    print("Sum of mean scores for A2C in", env, ":", a2c_sum)
    print("Sum of mean scores for SAC in", env, ":", sac_sum)
    print("Sum of mean scores for DQN in", env, ":", dqn_sum)
    plt.bar(labels, scores)
    plt.title("Sum of mean scores for each algorithm in " + env + " environment")
    plt.xlabel("Algorithm Name")
    plt.ylabel("Sum of mean scores")
    
    plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

environments = ["Bipedal Walker", "Mountain Car Continuous", "Lunar Lander", "Cart Pole", "Mountain Car", "Car Racing", "Pendulum", "Acrobot", "Bipedal Hardcore" ]


if not os.path.exists("plots"):
    os.mkdir("plots")

for env in environments:
    df = pd.read_csv("ALL DATA.csv")
    acrobot_data = df[df["Environment"] == env]
    ppo_sum = acrobot_data[acrobot_data["Algorithm Name"] == "PPO"]["Standard Deviation"].sum()
    td3_sum = acrobot_data[acrobot_data["Algorithm Name"] == "TD3"]["Standard Deviation"].sum()
    ddpg_sum = acrobot_data[acrobot_data["Algorithm Name"] == "DDPG"]["Standard Deviation"].sum()
    a2c_sum = acrobot_data[acrobot_data["Algorithm Name"] == "A2C"]["Standard Deviation"].sum()
    sac_sum = acrobot_data[acrobot_data["Algorithm Name"] == "SAC"]["Standard Deviation"].sum()
    dqn_sum = acrobot_data[acrobot_data["Algorithm Name"] == "DQN"]["Standard Deviation"].sum()
    labels = ["PPO", "TD3", "DDPG", "A2C", "SAC", "DQN"]
    scores = [ppo_sum, td3_sum, ddpg_sum, a2c_sum, sac_sum, dqn_sum]
    print("Sum of standard deviations for PPO in", env, ":", ppo_sum)
    print("Sum of standard deviations for TD3 in", env, ":", td3_sum)
    print("Sum of standard deviations for DDPG in", env, ":", ddpg_sum)
    print("Sum of standard deviations for A2C in", env, ":", a2c_sum)
    print("Sum of standard deviations for SAC in", env, ":", sac_sum)
    print("Sum of standard deviations for DQN in", env, ":", dqn_sum)
    total_sum = ppo_sum + td3_sum + ddpg_sum + a2c_sum + sac_sum + dqn_sum
    print("Total sum =" ,total_sum)
    plt.bar(labels, scores)
    plt.title("Sum of standard deviations for each algorithm in " + env + " environment")
    plt.xlabel("Algorithm Name")
    plt.ylabel("Sum of standard deviations")
    plt.savefig("plots/" + env + ".png")
    plt.close()
```
