```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("ALL DATA.csv")

acrobot_data = df[df["Environment"] == "Acrobot"]

algorithm_name = "PPO"  # Filter for PPO algorithm only
algorithm_data = acrobot_data[acrobot_data["Algorithm Name"] == algorithm_name]

# Calculate MPL
algorithm_data["MPL"] = algorithm_data["Mean Score"].diff() / algorithm_data["Training time"].diff()

# Define colors for each point
colors = ['red', 'green', 'blue', 'purple', 'orange', 'black', 'pink', 'brown']

# Define labels for each point
key_labels = ['10,000 episodes', '20,000 episodes', '50,000 episodes', '100,000 episodes', '500,000 episodes', '1 million episodes', '10 million episodes']

# Create scatter plot of MPL vs. training time
for i in range(len(key_labels)):
    plt.scatter([], [], c=colors[i], alpha=1, s=50, label=key_labels[i], edgecolors='black')

plt.scatter(algorithm_data["Training time"][1:], algorithm_data["MPL"][1:], c=colors[:len(algorithm_data)-1], s=50, edgecolors='black', linewidth=1.5, zorder=2)

# Add horizontal line for zero MPL
plt.axhline(y=0, color='gray', linestyle='--', linewidth=1)

# Add a line connecting the dots
plt.plot(algorithm_data["Training time"][1:], algorithm_data["MPL"][1:], linewidth=2, color='gray', zorder=1)

plt.xlabel("Training time (s)")
plt.ylabel("Marginal product of labor (MPL)")
plt.title("Acrobot: MPL vs. Training time (PPO only)")
plt.legend(title="Number of episodes")

plt.savefig("acrobot_mpl_vs_training_time_PPO.png")
plt.show()
```
