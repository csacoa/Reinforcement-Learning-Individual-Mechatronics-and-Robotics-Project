```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Only learning Rate is changing - Sheet1.csv")
df['Mean Score'] = pd.to_numeric(df['Mean Score'], errors='coerce')
mean_scores = df.groupby('Learning Rate')['Mean Score'].mean()
x = mean_scores.index
y = mean_scores.values
m, b = np.polyfit(x, y, 1)
plt.plot(x, y, 'o')
plt.plot(x, m*x + b)
plt.xlabel('Learning Rate')
plt.ylabel('Mean Score')
plt.title('Mean Score vs. Learning Rate')
#plt.savefig('mean_score_vs_learning_rate (100 episodes training).png')
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV data into a Pandas DataFrame
df = pd.read_csv("CartPole-v1_PPO Randomised Hyperparameters.csv")

# Convert the "Mean Score" column to numeric values
df['Mean Score'] = pd.to_numeric(df['Mean Score'], errors='coerce')

# Calculate the mean score for each learning rate
mean_scores = df.groupby('GAE Lambda')['Mean Score'].mean()

# Create a line plot of the mean scores vs. the learning rates, with a line of best fit
x = mean_scores.index
y = mean_scores.values
m, b = np.polyfit(x, y, 1)
plt.plot(x, y, 'o')
plt.plot(x, m*x + b)
plt.xlabel('GAE Lambda')
plt.ylabel('Mean Score')
plt.title('Mean Score vs. GAE Lambda')
plt.savefig('mean_score_vs_GAE_Lambda when all params are random (100 episodes training).png')
plt.show()


# Save the plot to a file
#plt.savefig('mean_score_vs_learning_rate (100 episodes training).png')
```
