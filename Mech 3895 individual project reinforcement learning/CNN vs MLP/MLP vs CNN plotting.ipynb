```python
import pandas as pd
import matplotlib.pyplot as plt

#for reading csv files
mlp_data = pd.read_csv("CarRacingMLPTest.csv")
cnn_data = pd.read_csv("CarRacingCnnTest.csv")


mlp_mean_scores = mlp_data["Mean Score"]
mlp_training_times = mlp_data["Training Time"]
cnn_mean_scores = cnn_data["Mean Score"]
cnn_training_times = cnn_data["Training Time"]

#this will plot the MLP vs CNN scores against eachother
fig, ax = plt.subplots()
ax.plot(mlp_training_times, mlp_mean_scores, color="blue", label="MLP")
ax.plot(cnn_training_times, cnn_mean_scores, color="red", label="CNN")

ax.set_title("Car Racing MLP and CNN Mean Score vs Training Time")
ax.set_xlabel("Training Time")
ax.set_ylabel("Mean Score")

ax.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


mlp_data = pd.read_csv("CarRacingMLPTest.csv")
cnn_data = pd.read_csv("CarRacingCnnTest.csv")
mlp_total_mean_score = mlp_data["Mean Score"].sum()
cnn_total_mean_score = cnn_data["Mean Score"].sum()
fig, ax = plt.subplots()
ax.bar(["MLP", "CNN"], [mlp_total_mean_score, cnn_total_mean_score], color=["blue", "red"])
ax.set_title("Total Mean Score for Car Racing MLP and CNN", fontsize = 16)
ax.set_xlabel("Model", fontsize = 16)
ax.set_ylabel("Total Mean Score", fontsize = 16)

#plt.savefig("Total Mean Score for Car Racing MLP and CNN.png")

plt.show()
```
