```python
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

df = pd.read_csv('CartPole-v1_PPO ALL HYPERPARAMETERS RANDOMISED AT 12500 EPISODES.csv')
X = df[['Learning Rate', 'Batch Size', 'Gamma', 'GAE Lambda']]
y = df['Mean Score']
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
regressor = LinearRegression()
regressor.fit(X_poly, y)
coeffs = regressor.coef_
intercept = regressor.intercept_
print('Coefficients: ', coeffs)
print('Intercept: ', intercept)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
df = pd.read_csv('CartPole-v1_PPO ALL HYPERPARAMETERS RANDOMISED AT 12500 EPISODES.csv')
X = df[['Learning Rate', 'Batch Size', 'Gamma', 'GAE Lambda']]
y = df['Mean Score']
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
model = LinearRegression().fit(X_poly, y)
print('Coefficients: ', model.coef_)
print('Intercept: ', model.intercept_)
scatter_matrix(df, figsize=(10, 10))
plt.show()
n = 100
xx = np.linspace(X.min().min(), X.max().max(), n)
yy = np.linspace(y.min(), y.max(), n)
X1, Y1 = np.meshgrid(xx, yy)
Z1 = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        x = np.array([X1[i,j], Y1[i,j], Y1[i,j]**2, X1[i,j]*Y1[i,j], X1[i,j]**2, X1[i,j]**2*Y1[i,j]])
        Z1[i,j] = model.predict(poly.transform([x]))
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X1, Y1, Z1, cmap='viridis')
ax.set_xlabel('Learning rate')
ax.set_ylabel('batch size')
ax.set_zlabel('Mean score')
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load data from CSV file
data = pd.read_csv('CartPole-v1_PPO ALL HYPERPARAMETERS RANDOMISED AT 12500 EPISODES.csv')

# Extract input and output variables
X = data[['Learning Rate']].values
y = data['Mean Score'].values

# Create linear regression model and fit to data
reg = LinearRegression()
reg.fit(X, y)

# Print the coefficients and intercept
print('Coefficients: ', reg.coef_)
print('Intercept: ', reg.intercept_)

# Plot the data and the regression line
plt.scatter(X, y, color='blue')
plt.plot(X, reg.predict(X), color='red', linewidth=2)
plt.title('Linear Regression of Mean Score vs Learning Rate')
plt.xlabel('Learning Rate')
plt.ylabel('Mean Score')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
data = pd.read_csv("CartPole-v1_PPO ALL HYPERPARAMETERS RANDOMISED AT 12500 EPISODES.csv")
X = data[['Learning Rate', 'Batch Size', 'Gamma', 'GAE Lambda']].values
Y = data['Mean Score'].values
degree = int(input("Enter the degree of polynomial: "))
poly = PolynomialFeatures(degree=degree)
X_poly = poly.fit_transform(X)
model = LinearRegression()
model.fit(X_poly, Y)
lr = float(input("Enter the Learning Rate: "))
bs = float(input("Enter the Batch Size: "))
g = float(input("Enter the Gamma: "))
gae = float(input("Enter the GAE Lambda: "))
x_input = poly.transform(np.array([[lr, bs, g, gae]])
y_pred = model.predict(x_input)

print("Predicted Mean Score:", y_pred[0])
```
